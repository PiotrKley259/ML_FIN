```python
import pandas as pd
import numpy as np
import itertools

data = pd.read_csv("JKP.csv", delimiter=",")
data['date'] = pd.to_datetime(data['date'], format='%Y-%m-%d')
data
```

```python
df = data.pivot(index='date', columns='name', values='ret').sort_index()
# delete dates before 1961-04-30, to match compustat data
df = df[df.index >= '1961-04-30']
df
```

```python
# print the columns names with missing values and the number of missing values
missing_values = df.isnull().sum()
missing_columns = missing_values[missing_values > 0]
print("Columns with missing values and their counts:")
print(missing_columns)
```

```python
# flag matrice qui marque les valeurs manquantes avec 1 et les valeurs non manquantes avec 0
flag_matrix = df.isna().astype(int)
```

```python
# fonction pour calculer les blocs de NaN dans les colonnes
def nan_blocks_info(series):
    """
    Retourne une liste des longueurs des blocs consécutifs de NaN dans la Series.
    """
    is_na = series.isna()
    run_lengths = []
    for val, group in itertools.groupby(is_na):
        length = sum(1 for _ in group)
        if val:  # si val == True => bloc de NaN
            run_lengths.append(length)
    return run_lengths

# résumé pour chaque colonne
summary = []
for col in df.columns:
    run_lengths = nan_blocks_info(df[col])
    summary.append({
        'factor': col,
        'num_blocks': len(run_lengths),
        'block_sizes': run_lengths
    })


summary_df = pd.DataFrame(summary)
summary_df
```

```python
# fonction pour calculer les blocs de NaN, leurs longueurs et les dates de début/fin
def nan_blocks_info(series):
    """
    Retourne une liste de dictionnaires pour les blocs de NaN dans la Series,
    avec 'start_date', 'end_date' et 'length' pour chaque bloc.
    """
    is_na = series.isna().tolist()
    dates = series.index.tolist()
    blocks = []
    
    i = 0
    n = len(series)
    while i < n:
        if is_na[i]:
            start_idx = i
            # Avancer jusqu'à la fin du bloc NaN
            while i < n and is_na[i]:
                i += 1
            end_idx = i - 1
            # Conserver le bloc
            blocks.append({
                'start_date': dates[start_idx],
                'end_date': dates[end_idx],
                'length': end_idx - start_idx + 1
            })
        else:
            i += 1
            
    return blocks


detailed_blocks = []
for col in df.columns:
    blocks = nan_blocks_info(df[col])
    for block in blocks:
        detailed_blocks.append({
            'factor': col,
            'start_date': block['start_date'],
            'end_date': block['end_date'],
            'length': block['length']
        })

detailed_df = pd.DataFrame(detailed_blocks)
print(detailed_df)
```

```python
# remplir par 0 tous les blocs qui commencent le 1961-04-30
target_start = pd.Timestamp('1961-04-30')
for factor in df.columns:
    blocks = nan_blocks_info(df[factor])
    for block in blocks:
        if block['start_date'] == target_start:
            df.loc[block['start_date']:block['end_date'], factor] = 0.0

# forward fill les NaN restants
df.fillna(method='ffill', inplace=True)


print("Nombre de NaN restant par colonne :")
df.isna().sum()
```

```python
df['year'] = df.index.year

# calculer une 'subperiod' 
start_year = df['year'].min()  
df['subperiod'] = ((df['year'] - start_year) // 5)


def label_subperiod(x):
    start = start_year + 5*x
    end = start + 4
    return f"{start}-{end}"
df['subperiod_label'] = df['subperiod'].map(label_subperiod)

# winsorisation (1e-99e percentiles), par subperiod et par facteur
df_winsor = df.copy()
factor_cols = df.columns.drop(['year', 'subperiod', 'subperiod_label'])

for sub_label, group in df_winsor.groupby('subperiod_label'):
    lower_bounds = group[factor_cols].quantile(0.01)
    upper_bounds = group[factor_cols].quantile(0.99)
    mask = df_winsor['subperiod_label'] == sub_label
    df_winsor.loc[mask, factor_cols] = df_winsor.loc[mask, factor_cols].clip(
        lower=lower_bounds, upper=upper_bounds, axis=1
    )

# retire les colonnes ajoutées
df_winsor = df_winsor.drop(columns=['year', 'subperiod', 'subperiod_label'])

df_winsor
```

```python
df_scaled = df_winsor.copy()

# colonne year pour découper en tranches de 5 ans
df_scaled['year'] = df_scaled.index.year
start_year = df_scaled['year'].min()
df_scaled['subperiod'] = (df_scaled['year'] - start_year) // 5

# pour chaque subperiod appliquer un z-score scaling
for sp, group in df_scaled.groupby('subperiod'):
    bloc = group[df_winsor.columns]  # uniquement les colonnes de facteurs
    mean = bloc.mean()
    std = bloc.std()
    scaled_block = (bloc - mean) / std
    # gérer les colonnes avec std = 0 (pas de variation), mettre 0
    scaled_block = scaled_block.fillna(0.0)
    # réinjecter dans df_scaled
    df_scaled.loc[bloc.index, bloc.columns] = scaled_block

# supprimer les colonnes 
df_scaled = df_scaled.drop(columns=['year', 'subperiod'])

df_scaled.head(10)
```

```python
# ajouer la matrice des flages à la fin de df_scaled
df_clean = pd.concat([df_scaled, flag_matrix], axis=1)
df_clean.to_csv("JKP_clean.csv", index=True)
```
